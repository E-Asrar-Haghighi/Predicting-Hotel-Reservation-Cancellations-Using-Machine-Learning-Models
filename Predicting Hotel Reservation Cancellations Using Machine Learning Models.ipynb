# **Project Summary**

This project aimed to predict hotel reservation cancellations using machine learning techniques, with a focus on understanding the factors influencing cancellations and optimizing prediction performance.

The dataset used in this project was obtained from **Kaggle**, a well-known platform for open datasets and machine learning projects. The dataset, titled **Hotel Reservations Classification Dataset**, is publicly available and structured as a **CSV file**, making it easy to load and process using Python's **pandas** library. Since it is publicly accessible, no additional data collection or web scraping was required.

The analysis revealed that certain factors, such as **lead time**, **online market segment**, **high average room price**, and **summer season**, significantly increased the likelihood of cancellations. In contrast, **special requests**, **offline bookings**, **winter season**, and **discounts** were associated with lower cancellation risks. 

We employed five machine learning models: **Logistic Regression**, **Decision Tree Classifier**, **Support Vector Machine (SVM)**, **Random Forest Classifier**, and **Gradient Boosting Classifier**. Among these, the **Gradient Boosting Classifier** consistently outperformed the others, achieving an **accuracy of 84.1%** and an **ROC-AUC score of 88.9%**, demonstrating its superior predictive capabilities.

To validate model performance, we applied **5-Fold Cross-Validation**, which confirmed the reliability and stability of the **Gradient Boosting Classifier** across different data subsets.

Addressing the issue of class imbalance was crucial, and the application of **SMOTE (Synthetic Minority Over-sampling Technique)** effectively increased **recall from 67.54% to 74.80%**, aligning with the project's goal of minimizing missed cancellations. Although there was a slight decline in precision, the improved recall was deemed a worthwhile trade-off.

Further optimization was achieved through **threshold adjustment**, where lowering the threshold to **0.4** enhanced **recall to 79.91%**, facilitating better detection of potential cancellations. This adjustment supports more proactive cancellation management, assisting hotels with **staffing, room allocation**, and **revenue planning**.

The study's insights have practical implications for hotel management. Strategies like **dynamic pricing**, **targeted marketing**, and **channel optimization** were identified as effective measures to reduce cancellation rates. Specifically:

- **Dynamic Pricing and Discounts**: Offering discounts during high-risk periods, such as summer, can reduce cancellations.
- **Targeted Marketing**: Engaging repeat guests and those with special requests can enhance booking reliability.
- **Channel Optimization**: Promoting offline and corporate bookings can further reduce cancellation rates.

The project also faced challenges such as **class imbalance**, which we addressed using **SMOTE** to generate synthetic samples for the minority class. This technique helped create a more balanced dataset, enabling the models to better capture patterns across different classes. Additionally, **feature engineering** and **normalization** played crucial roles in improving model performance by transforming raw data into meaningful inputs. To prevent overfitting, we applied **5-Fold Cross-Validation** to evaluate model performance across different data subsets. The **hyperparameter tuning** process involved adjusting parameters like **max_depth** and **min_samples_split**, which control model complexity and reduce the risk of overfitting. While this process required significant computational resources, the resulting improvements in recall and model generalization justified the effort.

In conclusion, the **SMOTE-balanced, hyperparameter-tuned Gradient Boosting Classifier** was identified as the most effective model for predicting hotel reservation cancellations. With high recall, stable performance, and insightful feature relevance, this model offers valuable predictive capabilities for hotel management.

**Future research** could explore real-time model retraining and integration with reservation systems to enhance predictive performance and adapt to evolving booking behaviors.

# 2. Dataset Overview
## **2.1 Description of the Dataset**  
The dataset used in this study is the **Hotel Reservations Classification Dataset**, sourced from Kaggle. It contains information on hotel bookings, customer details, and reservation statuses. The dataset is structured for a **binary classification task**, where the goal is to predict whether a reservation will be **canceled or not** based on various booking-related factors.  

The dataset consists of **multiple numerical and categorical features** that capture customer behavior, booking preferences, and reservation details. These factors influence the likelihood of cancellation and serve as input variables for machine learning models.  

With a sufficient number of observations and a balanced mix of categorical and numerical variables, this dataset provides a robust foundation for **exploratory data analysis (EDA) and machine learning modeling**.  


## **2.2 Key Features and Target Variable**  

### **- Target Variable**  
- **`booking_status`** (Categorical) → Indicates whether a booking was **Canceled** or **Not_Canceled**.  

### **- Features (Independent Variables)**  
The dataset includes the following features that could influence the cancellation of a hotel reservation:  

1. **Booking Details**  
   - `Booking_ID`: Unique identifier of each booking. 
   - `no_of_adults`: Number of adults in the booking.  
   - `no_of_children`: Number of children included in the booking.  
   - `no_of_weekend_nights`: Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel.  
   - `no_of_week_nights`: Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel. 
   - `type_of_meal_plan`: Type of meal plan booked by the customer.  
   - `room_type_reserved`: Type of room reserved by the customer. The values are ciphered (encoded) by INN Hotels.

2. **Guest Preferences and Special Requests**  
   - `required_car_parking_space`: Does the customer require a car parking space? (0 - No, 1- Yes).  
   - `market_segment_type`: Market segment designation.  
   - `repeated_guest`: Is the customer a repeated guest? (0 - No, 1- Yes).  

3. **Market and Payment Factors**  
   - `lead_time`: Number of days between the date of booking and the arrival date.  
   - `arrival_year`, `arrival_month`, `arrival_date`: Date of arrival.  
   - `deposit_type`: Type of deposit paid at the time of booking (e.g., No Deposit, Non-Refundable).  
   - `booking_changes`: Number of changes made to the reservation.  

4. **Reservation Confirmation and Customer Behavior** 
   - `no_of_previous_cancellations`: Number of previous bookings that were canceled by the customer prior to the current booking.
   - `no_of_previous_bookings_not_canceled`: Number of previous bookings not canceled by the customer prior to the current booking
   - `average_price_per_room`: Average price per day of the reservation; prices of the rooms are dynamic. (in euros).  
   - `no_of_special_requests`: Total number of special requests made by the customer (e.g. high floor, view from the room, etc).  

These features will be analyzed to determine their correlation with reservation cancellations. By applying **machine learning models**, we aim to identify which features are the strongest predictors of booking cancellations.

# 4. **Feature Engineering for Predicting Hotel Reservation Cancellations**
To improve the predictive performance of our machine learning model, we create additional features that could capture meaningful patterns related to cancellations.

## 4.1 **Temporal Features**

To capture seasonal variations that might influence cancellation behavior, temporal features were extracted from the `arrival_month` attribute. A custom `get_season` function was implemented to classify each month into one of four seasons: `Winter` (December, January, February), `Spring` (March, April, May), `Summer` (June, July, August), and `Fall` (September, October, November). This seasonal categorization helps the model consider the impact of peak and off-peak seasons on cancellation patterns, potentially enhancing its predictive accuracy.  
The distribution of reservations across seasons reveals that Fall accounted for the highest proportion of bookings at approximately 32.53%, followed by Summer at 27.67%, Spring at 22.58%, and Winter at 17.21%. This insight underscores the importance of seasonality in understanding customer booking behaviors and potential cancellations.

````python
  def get_season(month):
      if month in [12, 1, 2]: return "Winter"
      elif month in [3, 4, 5]: return "Spring"
      elif month in [6, 7, 8]: return "Summer"
      else: return "Fall"
  df['season'] = df['arrival_month'].apply(get_season)
````

## 4.2 **Lead Time Transformation**

To enhance the model’s ability to capture temporal patterns, the `lead_time` feature was transformed by categorizing it into meaningful segments. Using the `pd.cut` function, lead times were binned into four categories:
- **Very Short:** 0–7 days (last-minute bookings)
- **Short:** 8–30 days
- **Medium:** 31–90 days
- **Long:** 91+ days 

This transformation helps the model differentiate between short and long lead times, potentially improving predictions by introducing categorical distinctions based on lead time durations.

The distribution of bookings across these categories indicates that the majority of reservations fall into the `Medium` category at approximately 32.92%, followed by `Long` lead times at 29.27%, `Short` lead times at 22.06%, and `Very Short` lead times at 15.75%. This highlights that bookings made far in advance are more frequent and also more prone to cancellation, reinforcing the importance of this feature in the prediction model.

````python
df['lead_time_category'] = pd.cut(df['lead_time'], bins=[0, 7, 30, 90, 365], labels=['Very Short', 'Short', 'Medium', 'Long'])
````

````python
df['lead_time_category'].value_counts(normalize = True)
````

## 4.3 Customer Behavior Features

To capture the impact of a customer’s past booking behavior on the likelihood of cancellation, a `cancellation_ratio` feature was engineered. This ratio reflects the proportion of past bookings that resulted in cancellations for each customer, providing a personalized indicator of cancellation risk.

The calculated `cancellation_ratio` ranges from 0 to 0.9897, indicating that some customers have nearly all their previous bookings canceled, while others have minimal or no prior cancellations. Incorporating this behavioral feature allows the model to better assess individual customer tendencies, which is crucial for accurately predicting future cancellations.

````python
df['cancellation_ratio'] = df['no_of_previous_cancellations'] / (df['no_of_previous_cancellations'] + df['no_of_previous_bookings_not_canceled'] + 1)
````

````python
df['cancellation_ratio'].value_counts(normalize=True)
````

## 4.4 Special Request Count

The `no_of_special_requests` feature was transformed into a binary variable to indicate the level of customer engagement. Bookings with a high number of special requests were categorized separately to capture potential behavioral patterns linked to cancellations.

The distribution of this feature shows that approximately 97.38% of bookings had low or no special requests, while only 2.62% of bookings had a high number of special requests. This feature helps the model assess whether customers with more specific needs or preferences exhibit different cancellation behaviors compared to those with fewer or no special requests.

````python
df['high_special_requests'] = (df['no_of_special_requests'] > 2).astype(int)
````

````python
df['high_special_requests'].value_counts(normalize=True)
````

## 4.5 Pricing Features

To capture the impact of pricing on cancellation behavior, a `discount_applied` feature was created by comparing the room price of each booking with the average price for the same room type. Bookings priced below the average were marked as having a discount applied, while others were not.

The distribution indicates that approximately 52.95% of bookings received a discount, while 47.05% did not. This feature helps the model understand how pricing incentives influence customer decisions, including the likelihood of cancellations.

````python
df['discount_applied'] = (df['avg_price_per_room'] < df.groupby('room_type_reserved')['avg_price_per_room'].transform('mean')).astype(int)
````

````python
df['discount_applied'].value_counts(normalize=True)
````

## 4.6 Price per Night

To ensure that the cost of a booking is fairly represented regardless of the length of stay, a `price_per_night` feature was created by normalizing the `avg_price_per_room` based on the total number of nights (both weekend and weekday nights). This transformation ensures that the pricing information is consistent across bookings of varying lengths, allowing the model to better capture the relationship between nightly rates and cancellation behavior.

````python
df['price_per_night'] = df['avg_price_per_room'] / (df['no_of_weekend_nights'] + df['no_of_week_nights'] + 1)
````

## 4.7 Market & Customer Segmentation

To account for differences in cancellation behavior across various customer segments, a `segment_cancellation_rate` feature was created. This feature represents the historical cancellation rate for each `market_segment_type`, calculated as the proportion of canceled bookings within each segment. 

By incorporating segment-specific cancellation rates, the model can better capture distinct patterns of behavior associated with different market segments, enhancing its ability to predict cancellations more accurately.

````python
segment_cancellation_rate = df.groupby('market_segment_type')['booking_status'].apply(lambda x: (x == 'Canceled').mean())
df['segment_cancellation_rate'] = df['market_segment_type'].map(segment_cancellation_rate)
````

## 4.8 Booking Duration

The `total_nights` feature was created by summing the number of weekday and weekend nights for each booking. Booking duration can significantly influence cancellation behavior, as longer stays may either increase the risk of cancellation due to higher costs or reduce it due to more deliberate planning, depending on the customer type and booking conditions. This feature provides the model with additional context to better assess the impact of stay length on the likelihood of cancellations.

````python
df['total_nights'] = df['no_of_week_nights'] + df['no_of_weekend_nights']
````

### 6.3.3 Handling Skewness in the `no_of_previous_bookings_not_canceled` Feature  

#### **Observations on Data Distribution**  
The `no_of_previous_bookings_not_canceled` feature, representing the number of past bookings that were not canceled, exhibited a highly skewed distribution:

- A small proportion of guests had multiple past non-canceled bookings, with some reaching up to **58 successful past reservations**.

#### **Rationale for Feature Transformation**  
Using `no_of_previous_bookings_not_canceled` as a continuous numerical feature could introduce bias due to the large disparity in values. To mitigate this, a **categorical binning approach** was employed to group guests based on their booking history, reducing complexity while preserving meaningful information.

#### **Feature Binning Strategy**  
The feature was transformed into categorical groups to capture guest loyalty patterns:

| **Category**         | **Value Range**                                   | **Description**                                |
|----------------------|--------------------------------------------------|------------------------------------------------|
| **New Guest**        | `0`                                              | First-time guest, no previous successful bookings |
| **Occasional Guest** | `1-2`                                            | Successfully booked once or twice before       |
| **Returning Guest**  | `3-5`                                            | Moderate booking history                       |
| **Frequent Guest**   | `6-10`                                           | Several successful stays                       |
| **Loyal Guest**      | `>10`                                            | Long booking history, repeat customer          |

#### **Implementation and Encoding**  
- The `no_of_previous_bookings_not_canceled` feature was transformed using a custom binning function to assign each guest to one of the defined categories.
- The new categorical feature was then **one-hot encoded** for compatibility with machine learning algorithms.
- The original continuous feature was removed to prevent redundancy and improve model efficiency.

#### **Benefits of This Transformation**  
- **Reduced the impact of extreme values**, ensuring that the model is more robust and less biased.
- **Enhanced interpretability** by categorizing guests into meaningful loyalty groups.
- **Improved predictive power**, as cancellation behavior may differ significantly between new and returning guests.

````python
df['no_of_previous_bookings_not_canceled'].value_counts(normalize = True)
````

````python
# Define binning function for no_of_previous_bookings_not_canceled
def categorize_previous_bookings(x):
    if x == 0:
        return "New Guest"
    elif x <= 2:
        return "Occasional Guest"
    elif x <= 5:
        return "Returning Guest"
    elif x <= 10:
        return "Frequent Guest"
    else:
        return "Loyal Guest"

# Apply binning transformation
df['previous_booking_category'] = df['no_of_previous_bookings_not_canceled'].apply(categorize_previous_bookings)
````

# **13. Threshold Analysis for Enhanced Cancellation Detection**

## **13.1 Maximizing Cancellation Detection Through Threshold Adjustment**

The **performance of the Gradient Boosting model** was evaluated across **various classification thresholds** to determine the **optimal balance between precision and recall**. Since the **primary objective** is to **prevent cancellations**, **maximizing recall** is crucial — even at the expense of **precision**.


````python
import numpy as np
from sklearn.metrics import precision_recall_curve, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

# Predict probabilities for the test set
y_proba = grid_search_balanced.best_estimator_.predict_proba(X_test)[:, 1]

# Define threshold values to test
thresholds = np.arange(0.1, 0.9, 0.1)

# Evaluate performance at different thresholds
threshold_results = []

for threshold in thresholds:
    y_pred_thresh = (y_proba >= threshold).astype(int)
    acc = accuracy_score(y_test, y_pred_thresh)
    prec = precision_score(y_test, y_pred_thresh)
    rec = recall_score(y_test, y_pred_thresh)
    f1 = f1_score(y_test, y_pred_thresh)
    threshold_results.append((threshold, acc, prec, rec, f1))

# Convert results to a DataFrame for clarity
import pandas as pd
threshold_df = pd.DataFrame(threshold_results, columns=['Threshold', 'Accuracy', 'Precision', 'Recall', 'F1-Score'])

print("\nPerformance Metrics for Different Thresholds:")
print(threshold_df)
````
